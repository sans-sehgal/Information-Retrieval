```python
nodes = set()
edges = set()
venues = set()
for i in [0,1,2,3]:
    print(i)
    dblp = []
    with open('../dblp.v10/dblp-ref/dblp-ref-{}.json'.format(i) , 'r') as f: #change path to stored data 
        for line in f:
            dblp.append(line)
    for p in dblp:
        p = json.loads(p)
        # Venue (journal published in) acts as the query 
        v = p['venue']
        if v == '':
            continue
        id = p['id']
        # Number of citations act as relevance scores      
        citations = int((p['n_citation']/42508) * 4) + 1 # Dividing by the max number of citations to normalise values
        nodes.add((v , id, citations))
        venues.add(v)
        try:
            refs = p['references']
            for r in refs:
                edges.add((id,r))
        except Exception as e:
            pass
        
```

```python
nodes_new = sorted(nodes , key=lambda x: x[1])
```

```python
# encoding node ids to numbers since node2vec inputs numbers
i = 0
nodes_new = set()
dict = {}
for n in nodes:
    n = list(n)
    dict[n[1]] = i;
    n[1] = i
    n = tuple(n)
    nodes_new.add(n)
    i+=1
```

```python
nodes_new
```

```python
#encoding edges to numbers
edges_new = set()
count = 0
for e in edges:
    e = list(e)
    try:
        e[0] = dict[e[0]]
        e[1] = dict[e[1]]
        e = tuple(e)
        edges_new.add(e)
    except Exception as e:
        count+=1
        continue
print(count)
```

```python
#finally saving nodes, edges
with open('dblp_edges_test.p' , 'wb') as f:
    pickle.dump(edges_new , f)
with open('dblp_nodes_test.p' , 'wb') as f:
    pickle.dump(nodes_new , f)
```
